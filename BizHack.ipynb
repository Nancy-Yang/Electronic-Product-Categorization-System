```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def extract_features(input_url, typeName):
    # Specify url
    url = input_url

    # Package the request, send the request and catch the response: r
    r = requests.get(url)

    # Extracts the response as html: html_doc
    html_doc = r.text

    # create a BeautifulSoup object from the HTML: soup
    soup = BeautifulSoup(html_doc, "lxml")

    # Print the title of Guido's webpage

    table = soup.find_all("div", {"class": "productInfoContainer_2mCHp", 
                                 "data-automation":"pdp-product-specs-tab"})[0]

    features = []
    for item in table.find_all('div', {"class":"itemName_jGrp0"}):
        features.append(item.text)

    values = []
    for item in table.find_all('div', {"class":"itemValue_341-l"}):
        values.append(item.text)
    
    adict={}
    if (len(features) < 8):
        numOfFeatures = len(features)
    else:
        numOfFeatures = 8
    for i in range(numOfFeatures):
        adict[str(features[i])] = values[i]
    data = pd.DataFrame([adict], columns=adict.keys())
    data[typeName] = 1
    return data
```

```python
# extract links
def extract_links(input_url):
    # Specify url
    url = input_url

    # Package the request, send the request and catch the response: r
    r = requests.get(url)

        # Extracts the response as html: html_doc
    html_doc = r.text

        # create a BeautifulSoup object from the HTML: soup
    soup = BeautifulSoup(html_doc, "lxml")

        # Print the title of Guido's webpage
    links = []
    table = soup.find_all("div", {"class": "productsRow_DcaXn row_1Rbqw"})[0]
    for l in table.find_all('a'):
        links.append(l["href"])
    return links
```

```python
# 2-in-1
list_url_2in1=[]
for i in range(14):
    url = "https://www.bestbuy.ca/en-ca/category/2-in-1-laptops/36744?page=" + str(i)
    list_url_2in1.append(url)
```

```python
def iterate_pages(webpages):
    links = []
    for i in webpages:
        results = extract_links(i)
        links.extend(results)
    return links
```

```python
all_links = iterate_pages(list_url_2in1)
```

```python
for ind, l in enumerate(all_links):
    newdf = extract_features("https://www.bestbuy.ca" + l, "2-in-1")
    if (ind == 0):
        final_df_2in1 = newdf
    elif (ind < 60):
        final_df_2in1 = pd.merge(final_df_2in1, newdf, how='outer') 
    else:
        final_df_2in1 = pd.merge(final_df_2in1, newdf, how='left') 
final_df_2in1.head()
```
