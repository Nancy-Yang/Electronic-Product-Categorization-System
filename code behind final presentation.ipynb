```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.metrics import f1_score
```

```python
import numpy as np
```

```python
import pandas as pd
tablet=pd.read_csv('train data.csv')
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False, dtype=int)



demodict={'Display Type': 'TFT','Screen Size':10.1 ,'Native Screen Resolution':1920*1200 , 'Battery life(hr)': 13.0,'weight(g)':469.0 ,'height(cm)':25.62 }

demo_dataframe=pd.DataFrame(demodict,index=[0])

demo_dataframe=pd.concat((demo_dataframe, pd.DataFrame(ohe.fit_transform(demo_dataframe[['Display Type']]),
                            columns=ohe.categories_[0])),axis=1)
demo_dataframe=pd.concat((demo_dataframe, pd.DataFrame(ohe.fit_transform(demo_dataframe[['Native Screen Resolution']]),
                            columns=ohe.categories_[0])),axis=1)

Tablet=pd.concat((tablet, pd.DataFrame(ohe.fit_transform(tablet[['Display Type']]),
                            columns=ohe.categories_[0])),axis=1)
Tablet=pd.concat((Tablet, pd.DataFrame(ohe.fit_transform(Tablet[['Native Screen Resolution']]),
                            columns=ohe.categories_[0])),axis=1)
```

```python
Tablet.drop(columns=['Display Type','Native Screen Resolution'],inplace=True)
demo_dataframe.drop(columns=['Display Type','Native Screen Resolution'],inplace=True)
```

```python
iPad=Tablet[['iPad']]
iPadPro=Tablet[['iPad pro']]
iPadAir=Tablet[['iPad Air']]
iPadMini=Tablet[['iPad Mini']]
Android=Tablet[['Android ']]
Windows=Tablet[['Windows']]
X=pd.DataFrame(Tablet.drop(columns=['iPad','iPad pro','iPad Air','iPad Mini','Android ','Windows']))
col=X.columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X=scaler.fit_transform(X)
clf_knn = KNeighborsClassifier(n_neighbors=5)
clf_lr = LogisticRegression(class_weight='balanced')
clf_dt = DecisionTreeClassifier(min_samples_leaf=3, min_samples_split=9, random_state=500)
clf_vote1 = VotingClassifier(
    estimators=[('knn', clf_knn), ('lr', clf_lr), ('dt', clf_dt)]
)
clf_vote2 = VotingClassifier(
    estimators=[('knn', clf_knn), ('lr', clf_lr), ('dt', clf_dt)]
)
clf_vote3 = VotingClassifier(
    estimators=[('knn', clf_knn), ('lr', clf_lr), ('dt', clf_dt)]
)
clf_vote4 = VotingClassifier(
    estimators=[('knn', clf_knn), ('lr', clf_lr), ('dt', clf_dt)]
)
clf_vote5 = VotingClassifier(
    estimators=[('knn', clf_knn), ('lr', clf_lr), ('dt', clf_dt)]
)
clf_vote6 = VotingClassifier(
    estimators=[('knn', clf_knn), ('lr', clf_lr), ('dt', clf_dt)]
)
```

```python
clf_vote1.fit(X, iPad)
clf_vote2.fit(X, iPadPro)
clf_vote3.fit(X, iPadAir)
clf_vote4.fit(X, iPadMini)
clf_vote5.fit(X, Android)
clf_vote6.fit(X, Windows)
```
